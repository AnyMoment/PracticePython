### Find the outliers per Columns-Calculate the distance- Give the justification

```python
#import necessary data
from scipy import stats
import numpy as np
#remove the data which is not numerical
odf=df._get_numeric_data()
odf.head()
```

### Using z Score 

#### Resources

```python
#Resources
#https://machinelearningmastery.com/how-to-use-statistics-to-identify-outliers-in-data/
#https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba

#The commonly used threshold for outlier detection is usually a z-score of 3 or -3,
#i.e. being more than 3 standard deviations away from the mean 
```

```python
#using z score 
z = np.abs(stats.zscore(odf))
z
```

```python
z1=np.where(z>3)
z1
```

### Clip the outliers

```python
odf_z=odf[(z < 3).all(axis=1)]
odf_z.shape
```

```python
odf.shape

#we have removed 10 outlier data values
```

### Using IQR Score

#### Resources

```python
#https://haridas.in/outlier-removal-clustering.html
#https://www.researchgate.net/post/When_is_it_justifiable_to_exclude_outlier_data_points_from_statistical_analyses
#https://www.researchgate.net/post/Which_is_the_best_method_for_removing_outliers_in_a_data_set
#https://www.theanalysisfactor.com/outliers-to-drop-or-not-to-drop/
#https://www.kdnuggets.com/2017/02/removing-outliers-standard-deviation-python.html
```

```python
#IQR = Q3 - Q1

Q1=odf.quantile(0.25)
Q3=odf.quantile(0.75)
IQR=Q3-Q1
IQR
```

```python
# next we find the outliers ie outside the Q1 and Q3 range
print(odf<(Q1-1.5*IQR))|odf>(Q3+1.5*IQR)
#True values are the ones that outside the scope
```

### Clip the outliers using IQR

```python
odf_IQR = odf[~((odf < (Q1 - 1.5 * IQR)) |(odf > (Q3 + 1.5 * IQR))).any(axis=1)]
odf_IQR.shape
#Removed 67 Values
```
